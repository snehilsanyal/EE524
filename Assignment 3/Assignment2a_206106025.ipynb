# Q1) Rastrigin Function

```python
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np


#The X and Y axes
x = np.linspace(-5.12, 5.12, 300)    
y = np.linspace(-5.12, 5.12, 300)    
x1, x2 = np.meshgrid(x,y)

#Rastrigin Function 
#2 Dimensional Function
A=10;n=1
term1a= A*1
term2a=x**2
term3a=A*np.cos(2*np.pi*x)
fplane1=term1a+term2a-term3a
#3D function
term1b= A*n
term2b=x1**2+x2**2
term3b=A*np.cos(2*np.pi*x1)+A*np.cos(2*np.pi*x2)
fplane2=term1b+term2b-term3b

#Visualizing the function
colormap1=cm.magma_r

#Figure 0
fig0 = plt.figure()
ax0 = fig0.gca(projection='3d')
im0 = ax0.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
fig0.colorbar(im0,orientation='horizontal')
ax0.set_xlabel("X-Axes")
ax0.set_ylabel("Y-Axes")
ax0.set_zlabel("Z-Axis")
ax0.set_title("Rastrigin Function")



#Figure 1
fig= plt.figure(figsize=plt.figaspect(0.5))

#Figure 1.1
ax = fig.add_subplot(231, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(0,60)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")
ax.set_title("Rastrigin Function")


#Figure 1.2
ax = fig.add_subplot(232)
sd=plt.imshow(fplane2)
fig.colorbar(sd)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_title("Rastrigin Function 2D View")

#Figure 1.3
ax = fig.add_subplot(233, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax.view_init(30,90)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")
ax.set_title("Side View")

#Figure 1.4
ax = fig.add_subplot(234, projection='3d')
im = ax.plot_surface(x1, x2, fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(30,45)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")

#Figure 1.5
ax = fig.add_subplot(235, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax.view_init(30,70)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")

#Figure 1.6
ax = fig.add_subplot(236, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(45,90)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")

#ColorBar to the plots
fig.tight_layout()
fig.subplots_adjust(bottom=0.21)
cax = fig.add_axes([0.05, 0.1, 0.91, 0.03])
fig.colorbar(im, cax=cax,orientation='horizontal')
plt.show()

#Figure 2, 2D function 
fig1=plt.figure()
plt.plot(x,fplane1)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Rastrigin function 1D form")
plt.show()



#Figure 3
fig2= plt.figure(figsize=plt.figaspect(0.5))
xangle=60
yangle=260

#Figure 3.1
ax1 = fig2.add_subplot(231, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-5,5)
ax1.set_ylim(-5,5)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")
ax1.set_title("Rastrigin - Zoomed Out view")

#Figure 3.2
ax1 = fig2.add_subplot(232)
sd=plt.imshow(fplane2)
fig1.colorbar(sd)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_title("Rastrigin Function 2D View")

#Figure 3.3
ax1 = fig2.add_subplot(233, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-4,2)
ax1.set_ylim(-4,2)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")
ax1.set_title("Global Minimum")

#Figure 3.4
ax1 = fig2.add_subplot(234, projection='3d')
im = ax1.plot_surface(x1, x2, fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")

#Figure 3.5
ax1 = fig2.add_subplot(235, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-3,1.5)
ax1.set_ylim(-3,1.5)

#Figure 3.6
ax1 = fig2.add_subplot(236, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-2.5,1)
ax1.set_ylim(-2.5,1)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")
ax1.set_title("The path to minimum")

#ColorBar to the plots
fig1.tight_layout()
fig1.subplots_adjust(bottom=0.21)
cax1 = fig1.add_axes([0.05, 0.1, 0.91, 0.03])
fig1.colorbar(im, cax=cax1,orientation='horizontal')
plt.show()
```

# Q2) Ackley's Function

```python
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np


# the X and Y axes
x = np.linspace(-32.76, 32.76, 300)    
y = np.linspace(-32.76, 32.76, 300)    
x1, x2 = np.meshgrid(x, y)

#dimesions (n+1)
n=2

# The Ackley's Function
term1=-0.2*np.sqrt(1/n*np.square(x1)+1/n*np.square(x2))
term2=1/n*(np.cos(2*np.pi*x1)+np.cos(2*np.pi*x2))
fplane2=-20*np.exp(term1)-np.exp(term2)+20+np.e

# Visualizing the function
colormap1=cm.rainbow

#Figure 0
fig0 = plt.figure()
ax0 = fig0.gca(projection='3d')
im0 = ax0.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
fig0.colorbar(im0,orientation='horizontal')
ax0.set_xlabel("X-Axes")
ax0.set_ylabel("Y-Axes")
ax0.set_zlabel("Z-Axis")
ax0.set_title("Akcley's Function")

#Figure 1
fig= plt.figure(figsize=plt.figaspect(0.50))

#Figure 1.1
ax = fig.add_subplot(231, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(0,90)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")
ax.set_title("Ackley's Function 3D")

#Figure 1.2
ax = fig.add_subplot(232)
sd=plt.imshow(fplane2)
fig.colorbar(sd)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_title("Ackley's Function 2D View")

#Figure 1.3
ax = fig.add_subplot(233, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax.view_init(-30,170)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")


#Figure 1.4
ax = fig.add_subplot(235, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax.view_init(45,100)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")


#Figure 1.5
ax = fig.add_subplot(236, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(60,20)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")
ax.set_title("Central Hole")
             
#Figure 1.6
ax = fig.add_subplot(234, projection='3d')
im = ax.plot_surface(x1, x2, fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(30,90)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")
ax.set_title("Global Minimum")
             
#ColorBar to the plots
fig.tight_layout()
fig.subplots_adjust(bottom=0.21)
cax = fig.add_axes([0.05, 0.1, 0.91, 0.03])
fig.colorbar(im, cax=cax,orientation='horizontal')
plt.show()

#Figure properties
fig1= plt.figure(figsize=plt.figaspect(0.5))
xangle=-60
yangle=140

#Figure 2.1
ax1 = fig1.add_subplot(231, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-40,40)
ax1.set_ylim(-40,40)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")

#Figure 2.2
ax1 = fig1.add_subplot(232)
sd=plt.imshow(fplane2)
fig.colorbar(sd)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_title("Ackley's Function")

#Figure 2.3
ax1 = fig1.add_subplot(233, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-5,5)
ax1.set_ylim(-5,5)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")
ax1.set_title("Central well")

#Figure 2.4
ax1 = fig1.add_subplot(235, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax1.set_xlim(-3,3)
ax1.set_ylim(-3,3)
ax1.view_init(xangle,yangle)
ax1.set_title("Global Minimum")

#Figure 2.5
ax1 = fig1.add_subplot(236, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.set_xlim(-20,20)
ax1.set_ylim(-20,20)
ax1.view_init(xangle,yangle)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")


#Figure 2.6
ax1 = fig1.add_subplot(234, projection='3d')
im = ax1.plot_surface(x1, x2, fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")

#ColorBar for the plots 
fig1.tight_layout()
fig1.subplots_adjust(bottom=0.21)
cax1 = fig1.add_axes([0.05, 0.1, 0.91, 0.03])
fig1.colorbar(im, cax=cax1,orientation='horizontal')
plt.show()
```

# Q4) Rosenbrock Function

```python

from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np


#The X and Y axes
x = np.linspace(-1.2, 1.2, 300)
y = np.linspace(-1.2, 1.2, 300)
x1, x2 = np.meshgrid(x,y)

#Rosenbrock Function
fd=2*x1*x1+x2*x2-2*x1*(1+x2)+1
fg=100*(x1*x1-x2)**2+(1-x1)**2

fplane2=fd+fg

#Visualizing the function
colormap1=cm.magma_r

#Figure 0
fig0 = plt.figure()
ax0 = fig0.gca(projection='3d')
im0 = ax0.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
fig0.colorbar(im0,orientation='horizontal')
ax0.set_xlabel("X-Axes")
ax0.set_ylabel("Y-Axes")
ax0.set_zlabel("Z-Axis")
ax0.set_title("Rosenbrock Function")

#Figure 1
fig= plt.figure(figsize=plt.figaspect(0.5))

#Figure 1.1
ax = fig.add_subplot(231, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(0,90)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")
ax.set_title("Rosenbrock Function 3D -Side View")

#Figure 1.2
ax = fig.add_subplot(232)
sd=plt.imshow(fplane2)
fig.colorbar(sd)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_title("Rosenbrock Function")

#Figure 1.3
ax = fig.add_subplot(233, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax.view_init(60,170)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")
ax.set_title("Global Minimum")

#Figure 1.4
ax = fig.add_subplot(234, projection='3d')
im = ax.plot_surface(x1, x2, fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(60,45)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")


#Figure 1.5
ax = fig.add_subplot(235, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax.view_init(30,130)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")

#Figure 1.6
ax = fig.add_subplot(236, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(45,20)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")

#ColorBar to the figure
fig.tight_layout()
fig.subplots_adjust(bottom=0.21)
cax = fig.add_axes([0.05, 0.1, 0.91, 0.03])
fig.colorbar(im, cax=cax,orientation='horizontal')
plt.show()

#Figure 2
fig1= plt.figure(figsize=plt.figaspect(0.5))
xangle=60
yangle=140

#Figure 2.1
ax1 = fig1.add_subplot(231, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-1.5,1.5)
ax1.set_ylim(-1.5,1.5)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")
ax1.set_title("Rosenbrock Function - Zoomed Out view")

#Figure 2.2
ax1 = fig1.add_subplot(232)
sd=plt.imshow(fplane2)
fig1.colorbar(sd)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_title("Rosenbrock Function")

#Figure 2.3
ax1 = fig1.add_subplot(233, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-.85,.85)
ax1.set_ylim(-.85,.85)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")
ax1.set_title("Global Minimum")

#Figure 2.4
ax1 = fig1.add_subplot(234, projection='3d')
im = ax1.plot_surface(x1, x2, fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")
ax1.set_title("The path to minimum")

#Figure 2.5
ax1 = fig1.add_subplot(235, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-1,1)
ax1.set_ylim(-1,1)

#Figure 2.6
ax1 = fig1.add_subplot(236, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-.6,.6)
ax1.set_ylim(-.6,.6)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")

#ColorBar for the plots
fig1.tight_layout()
fig1.subplots_adjust(bottom=0.21)
cax1 = fig1.add_axes([0.05, 0.1, 0.91, 0.03])
fig1.colorbar(im, cax=cax1,orientation='horizontal')
plt.show()
```

# Q5) McCormick Function

```python
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np


#The X and Y axes
x = np.linspace(-1.5, 4, 300)    
y = np.linspace(-3, 4, 300)    
x1, x2 = np.meshgrid(y,x)

#McCormick Function
term1=np.sin(x1+x2)
term2=(x1-x2)**2
term3=1.5*x1
term4=2.5*x2+1
fplane2=term1+term2+term3+term4

#Visualizing the function
colormap1=cm.viridis



#Figure 0
fig0 = plt.figure()
ax0 = fig0.gca(projection='3d')
im0 = ax0.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
fig0.colorbar(im0,orientation='horizontal')
ax0.set_xlabel("X-Axes")
ax0.set_ylabel("Y-Axes")
ax0.set_zlabel("Z-Axis")
ax0.set_title("McCormick Function")




#Figure 1
fig= plt.figure(figsize=plt.figaspect(0.5))

#Figure 1.1
ax = fig.add_subplot(231, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(0,290)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")
ax.set_title("McCormick Function 3D -Side View")

#Figure 1.2

ax = fig.add_subplot(232)
sd=plt.imshow(fplane2)
fig.colorbar(sd)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_title("McCormick Function")

#Figure 1.3
ax = fig.add_subplot(233, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax.view_init(60,170)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")
ax.set_title("The path to global minimum")

#Figure 1.4
ax = fig.add_subplot(234, projection='3d')
im = ax.plot_surface(x1, x2, fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(60,260)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")

#Figure 1.5
ax = fig.add_subplot(235, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax.view_init(30,130)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")

#Figure 1.6
ax = fig.add_subplot(236, projection='3d')
im = ax.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax.view_init(-45,120)
ax.set_xlabel("X-Axes")
ax.set_ylabel("Y-Axes")
ax.set_zlabel("Z-Axis")

#ColorBar to the plots
fig.tight_layout()
fig.subplots_adjust(bottom=0.21)
cax = fig.add_axes([0.05, 0.1, 0.91, 0.03])
fig.colorbar(im, cax=cax,orientation='horizontal')
plt.show()



#Figure 2
fig1= plt.figure(figsize=plt.figaspect(0.5))
xangle=60
yangle=260

#Figure 2.1
ax1 = fig1.add_subplot(231, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-5,5)
ax1.set_ylim(-5,5)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")
ax1.set_title("McCormick Function - Zoomed Out view")

#Figure 2.2
ax1 = fig1.add_subplot(232)
sd=plt.imshow(fplane2)
fig1.colorbar(sd)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_title("McCormick Function 3D View")

#Figure 2.3
ax1 = fig1.add_subplot(233, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-4,2)
ax1.set_ylim(-4,2)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")
ax1.set_title("Global Minimum")

#Figure 2.4
ax1 = fig1.add_subplot(234, projection='3d')
im = ax1.plot_surface(x1, x2, fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")
ax1.set_title("The path to minimum")

#Figure 2.5
ax1 = fig1.add_subplot(235, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1, linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-3,1.5)
ax1.set_ylim(-3,1.5)

#Figure 2.6
ax1 = fig1.add_subplot(236, projection='3d')
im = ax1.plot_surface(x1,x2,fplane2, rstride=1, cstride=1, cmap=colormap1,linewidth=0, antialiased=False)
ax1.view_init(xangle,yangle)
ax1.set_xlim(-2.5,1)
ax1.set_ylim(-2.5,1)
ax1.set_xlabel("X-Axes")
ax1.set_ylabel("Y-Axes")
ax1.set_zlabel("Z-Axis")

#ColorBar for the plots
fig1.tight_layout()
fig1.subplots_adjust(bottom=0.21)
cax1 = fig1.add_axes([0.05, 0.1, 0.91, 0.03])
fig1.colorbar(im, cax=cax1,orientation='horizontal')
plt.show()
```

In my understanding, their main difference is their minima

Their main difference lies in this nature of minima of these functions, which comes for use in many different problems. This nature of minima determines the nature of this function, whether the starting point influence the optimization or whether it will encounter many different local minimum before coming to the global minimum.
