```
# function for finding the factorial of a number.

def factorial_of_Num(num):

  fact = 1

  # check if the number is negative, positive or zero
  if num > 0:
    for i in range(1,num + 1):
        fact = fact * i
    print("The factorial of a given number is: \n{}\n" .format(fact))
  # Factorial of Num '0' is 1
  elif num == 0:
    print("The factorial of 0 is 1")
  # Factorial of Negative number doesn't exist
  else:
    print("Factorial of negative numbers does not exist.")


# Calling our Function
factorial_of_Num(7)
```

```
# Python Program to find the L.C.M. of two input number

def LCM(a, b):
   # choose the max from the numbers
   if a > b:
       max = a
   else:
       max = b

   while(True):
       if ((max % a == 0) and (max % b == 0)):
           lcm = max
           break
       max += 1

   return lcm
# Calling the LCM function
LCM(17,3)
```

```
# Function for Bubble Sorting

def Dec_Sort(list_of_num): 
    n = len(list_of_num) 
  
    # Traverse through all array elements 
    for i in range(n-1): 
  
        for j in range(0, n-i-1): 
  
            # traverse the array from 0 to n-i-1 
            # Swap if the element found is greater 
            # than the next element 
            if list_of_num[j] < list_of_num[j+1] : 
                list_of_num[j], list_of_num[j+1] = list_of_num[j+1], list_of_num[j]
  

# Creating a list of 15 random numbers and calling our function for sorting.
import random

list_of_num = random.sample(range(1, 50), 15)
print("Random number list: {}" .format((list_of_num)))

# Calling our function
Dec_Sort(list_of_num) 
  
print("Sorted array:")
for i in range(len(list_of_num)): 
    print("%d" %list_of_num[i]), 
```

```
# Imported math lib for math compution like add, sub, etc,.
import math
from math import sqrt

# Defining our class for basic operation of complex numbers.
class Complex(object):
  # _init_ for  an object creation from a class and to initialize the attributes of the class
  def __init__(self, real, imag=0.0):
      self.real = real
      self.imag = imag
      # Formats our results
      print(self.real + self.imag)

  def __abs__(self):
      print('\nAbsolute Value of the complex number:')
      new = (self.real**2 + (self.imag**2)*-1)
      return Complex(sqrt(new.real))

  def __add__(self, other):
      print('\nSum of the complex numbers:')
      return Complex(self.real + other.real, self.imag + other.imag)

  def __sub__(self, other):
      print('\nDifference of the complex number:')
      return Complex(self.real - other.real, self.imag - other.imag)
  
  def __mul__(self, other):
      print('\nProduct of the complex number:')
      return Complex((self.real * other.real) - (self.imag * other.imag),
          (self.imag * other.real) + (self.real * other.imag))

  def __truediv__(self, other):
      print('\nQuotient:')
      r = (other.real**2 + other.imag**2)
      return Complex((self.real*other.real - self.imag*other.imag)/r,
          (self.imag*other.real + self.real*other.imag)/r)


i = Complex(2, 10j)
k = Complex(3, 5j)

abs(i)
abs(k)
i + k
i - k
i * k
i / k
```
