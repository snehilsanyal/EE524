```python
from mpl_toolkits import mplot3d
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 2  #slope of the line 
b = 3  #y-axis intercept
```

```python
x = np.linspace(-10,10,100)
y = a*x + b
plt.plot(x,y,'-r', label= 'y = 2x + 3',c = 'black')
plt.title('graph of y=2x+3')
plt.xlabel('x',color='#1C2833')
plt.ylabel('y',color='#1C2833')
y1 = a*x + b + np.random.normal(0,1,100)
plt.scatter(x,y1,c='red')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
 x= np.linspace(-10,10,100,endpoint=True)
 a,b = np.meshgrid(np.arange(-10,11,0.1),np.arange(-10,11,0.1))
def  f(a, b):
    err = 0
    sigma= 5
    z= np.random.normal(0,sigma,1)    #generation of random point
    for i in range(len(x)):
        yi=a*x[i]+b
        yih=2*x[i]+3+z
        err = np.square(yi-yih)+err
    return err/10  
error=f(a,b)
fig=plt.figure(figsize=(10,10))
ax=plt.axes(projection='3d')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('error')
ax.plot_surface(a,b,error,cmap='viridis')
plt.show()
```
