```
a = np.arange(-10, 10, 0.1)
b = np.arange(-10, 10, 0.1)
E = np.empty((len(a),len(b)), dtype=float)

#(a) loop for considering all the combination of a,b
for i in range(len(a)):
  for j in range(len(b)):
    y = a[i]*x+b[j]
    #(b) Error btw the noisy z line and selected possible y line
    error = z - y
    sum = 0
    #(c) Loop through the array  to calculate Mean square error   
    for k in range(0, len(error)):
      E[i][j] = E[i][j] + ((error[k])*(error[k]))
    #(d) Stores all the possible MSE for different a,b values 
    E[i][j] = E[i][j]/len(error)
```

```
#(e) ploting 
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(1, 1, 1, projection='3d')

x = np.linspace(-10, 10,200)
y = np.linspace(-10, 10,200)
Z = E
X, Y = np.meshgrid(x, y)
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm,linewidth=0, antialiased=False)
fig.colorbar(surf)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```
