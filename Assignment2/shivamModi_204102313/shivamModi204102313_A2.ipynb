1. Generate a set of points around a line y = ax + b

(a) Choose a = 2 and b = 3

(b) Select the range for x as [âˆ’10, 10] and generate n = 100 values for x in that interval.

(c) Compute the values of y for each x as yi = 2xi + 3.

(d) Plot the line y = 2x + 3 in black color.

(e) Generate a set of n points around the line using the equation yi = 2xi + 3 + sigma N(0, 1) (1)where sigma is the standard deviation and N(0, 1) is the zero-mean unity-variance normal distribution

(f) Show the scatter plot of these noisy points (in red color) on the samegraph generated in step (d).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
n=100
a=2
b=3
x=np.linspace(-10,10,n)
print(x)
y=a*x+b
print(y)
plt.plot(x,y,color="k")

var=1
sigma= var**.5
mean=0
N=np.random.normal(mean,sigma,n)
y_i=a*x+b+sigma*N
print(y_i)
plt.scatter(x,y_i,color="r")
```

2. Plot the average error surface E for different values of a and b in the interval of [âˆ’10 : 0.1 : 10].

(a) Vary both a and b in steps of 0.1 in the interval [âˆ’10, 10]

(b) Compute the element-wise error as ei = yi âˆ’ Ë†yi where Ë†yi = axi + b and yi is computed using equation 1

(c) Compute the average error asE =1nnXi=iei2 (2)

(d) Compute the average error values for all combinations of a and b.

(e) Plot the error surface with the values of a along x-axis, that of b along y-axis and E along z-axis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



fig=plt.figure()
ax=fig.add_subplot(projection="3d")

arr=np.array([])       #this array is reuired to append upcoming values
x=np.linspace(-10,10,100)
a=np.linspace(-10,10,200)
b=np.linspace(-10,10,200)
n=100

#part of Q1
var=1
sigma= var**.5
grad=2
interception=3
mean=0
N=np.random.normal(mean,sigma,n)
y_i=grad*x+interception+sigma*N


p=len(a)
q=len(b)
#main logic
for i in range(p):
    for j in range(q):
        y2=i*x+j+sigma*N
        ER_sq=(y_i- y2)**2
        addition=ER_sq.sum()
        avg_error=addition/len(y2)
        arr=np.append(arr,avg_error)

#printing
print(len(arr))
arr=np.reshape(arr,[p,q])    #always pass the name and desired shape inside also
a,b=np.meshgrid(a,b)

ax.plot_surface(a,b,arr)


ax.set_xlabel("variable a")
ax.set_ylabel("variable b")
ax.set_zlabel("Error")
plt.show()
           
```
