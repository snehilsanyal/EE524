```
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import math
```

**Rastrigin's Optimization Function in 2D**

General range for x in Rastrigin's optimization is [-5.12,5.12]


```
#2-D Rastrigin Optimization Function
n = 1
x = np.linspace(-5.12,5.12,100)
ras_fun_2 = 10*n + (np.square(x)- 10*np.cos(2*math.pi*x))
plt.plot(x,ras_fun_2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

**Rastrigin's Optimization Function in 3D**

General range for x1 and x2 in Rastrigin's optimization is [-5.12,5.12]

```
#3-D Rastrigin Optimization Function
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
n = 2
x = np.linspace(-5.12,5.12,100)
y = np.linspace(-5.12,5.12,50)
x1,x2 = np.meshgrid(x,y)
ras_fun_3= 10*n + (np.square(x1) - 10*np.cos(2*math.pi*x1)) + (np.square(x2) - 10*np.cos(2*math.pi*x2))
my_cmap=plt.get_cmap('gist_earth')
ax.plot_surface(x1,x2,ras_fun_3,cmap = my_cmap,edgecolor = 'none')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

**Ackley's Optimization Function in 2D**

General range for x in Ackley's optimization is [-32.768,32.768]

```
#2-D Ackley's Optimization function
n = 1
x = np.linspace(-32.768,32.768,200)
ack_fun_2= -20*np.exp(-0.2*np.sqrt(np.square(x)/n)) - np.exp((np.cos(2*math.pi*x))/n) + 20 + math.exp(1)
plt.plot(x,ack_fun_2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

**Ackley's Optimization Function in 3D**

General range for x1 and x2 in Ackley's optimization is [-32.768,32.768]

```
#3-D Ackley's Optimization function
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
n = 2
x = np.linspace(-32.768,32.768,100)
y = np.linspace(-32.768,32.768,200)
x1,x2 = np.meshgrid(x,y)
ack_fun_3 = -20*np.exp(-0.2*np.sqrt((np.square(x1)+np.square(x2)/n))) - np.exp((np.cos(2*math.pi*x1)+np.cos(2*math.pi*x2))/n) + 20 + math.exp(1)
my_cmap=plt.get_cmap('gist_earth')
ax.plot_surface(x1,x2,ack_fun_3,cmap = my_cmap,edgecolor = 'none')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

**Sphere Optimization Function**

General range for x1 and x2 in Sphere optimization is [-5.12,5.12]

```
#Sphere function
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
n = 2
x = np.linspace(-5.12,5.12,100)
y = np.linspace(-5.12,5.12,100)
x1,x2 = np.meshgrid(x,y)
sph_fun = np.square(x1) + np.square(x2)
my_cmap=plt.get_cmap('gist_earth')
ax.plot_surface(x1,x2,sph_fun,cmap = my_cmap, edgecolor = 'none')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

**Rosenbrock Optimization Function**

General range for x1 and x2 in Rosenbrock's Optimization is [-5,10] but it can also be restricted to [-2.048,2.048]

```
#Rosenbrock Function
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
x1 = np.linspace(-5,10,100)
x2 = np.linspace(-5,10,200)
x1,x2 = np.meshgrid(x1,x2)
rbk_fun = 100*np.square(np.square(x1)-x2) + np.square(1 - x1)
my_cmap=plt.get_cmap('gist_earth')
ax.plot_surface(x1,x2,rbk_fun,cmap = my_cmap,edgecolor = 'none')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

**McCormick's Optimization Function**

General range for x1 in McCormick's Optimization is [-1.5,4] and for x2 is [-3,4]

```
#McCormick Function
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
x = np.linspace(-1.5,4,100) # Generally range of x for McCormick function is [-1.5,4]
y = np.linspace(-3,4,200) # Generally range of y for McCormick function is [-3,4]
x,y = np.meshgrid(x,y)
mcc_fun = np.sin(x + y) + np.square(x-y) -1.5*x + 2.5*y + 1 
my_cmap=plt.get_cmap('gist_earth')
ax.plot_surface(x,y,mcc_fun,cmap = my_cmap,edgecolor = 'none')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

**Matyas Optimization Function**

General range for x1 and x2 in Matyas's Optimization is [-10,10]

```
#Matyas Function
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
x,y = np.meshgrid(x,y)
maty_fun = 0.26*(np.square(x)+np.square(y)) - 0.48*x*y
my_cmap=plt.get_cmap('gist_earth')
ax.plot_surface(x,y,maty_fun,cmap = my_cmap,edgecolor = 'none')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```
