```python
#question-1

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np

x1 = np.linspace(-10,10,100)
y = 10*1+(x1**2-10*np.cos(2*np.pi*x1))
plt.plot(x1,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Rastrigin function 2-D plot')
plt.show()

x1 = np.linspace(-10,10,100) 
x2 = np.linspace(-10,10,100) 
x1, x2 = np.meshgrid(x1, x2)
z =10*2+(x1**2-10*np.cos(2*np.pi*x1))+(x2**2-10*np.cos(2*np.pi*x2))
fig = plt.figure()
ax1 = fig.gca(projection='3d')
ax1.set_title('Rastrigin function 3-D plot')
ax1.plot_surface(x1,x2,z,cmap=cm.plasma)  
plt.show()
```

```python
#question-2

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

x = np.linspace(-10,10,100)
z =-20*np.exp(-0.2*np.sqrt(x**2/1)) -np.exp(np.cos(2*np.pi*x)/1) + np.exp(1)+ 20
plt.plot(x,z)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Ackleys function 2-D plot')
plt.show()

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
x,y = np.meshgrid(x,y)
z1=0.2+np.exp(1)-20*np.exp(-0.2*np.sqrt(x*x + y*y)/2)-np.exp((np.cos(2*np.pi*x)+np.cos(2*np.pi*y))/2)
fig = plt.figure()
ax1 = fig.gca(projection='3d')
ax1.set_title('Ackleys function 3-D plot')
ax1.plot_surface(x, y,z1,cmap=cm.coolwarm)  
plt.show()
          
```

```python
#question3

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plot
from matplotlib import cm
import numpy as np

x1 = np.linspace(-10,10,100) 
x2 = np.linspace(-10,10,100) 
x1, x2 = np.meshgrid(x1, x2)
z = x1**2 +x2**2
fig = plt.figure()
ax1 = fig.gca(projection='3d')
ax1.set_title(' Sphere function')
ax1.plot_surface(x1,x2,z,cmap=cm.rainbow)
plt.show()
```

```python
#question-4

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plot
from matplotlib import cm
import numpy as np

x1 = np.linspace(-10,10,100) 
x2 = np.linspace(-10,10,100) 
x1, x2 = np.meshgrid(x1, x2)
z = 100*(x1**2-x2)**2 +(1-x1)**2
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_title(' Rosenbrock function.')
ax.plot_surface(x1, x2, z,cmap=cm.hsv)
plt.show()
```

```python
#question-5

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plot
from matplotlib import cm
import numpy as np

x = np.linspace(-10,10,100) 
y = np.linspace(-10,10,100) 
x,y = np.meshgrid(x,y)
z = np.sin(x+y)+(x-y)**2-1.5*x+2.5*y+1
fig = plot.figure()
ax1 = fig.gca(projection='3d')
ax1.set_title(' McCormick function')
ax1.plot_surface(x,y,z,cmap=cm.jet)
plt.show()
```

```python
#question-6

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plot
from matplotlib import cm
import numpy as np

x = np.linspace(-10,10,100) 
y = np.linspace(-10,10,100) 
x,y = np.meshgrid(x,y)
z = 0.26*(x**2+y**2) - (0.48*x*y)
fig = plt.figure()
ax1 = fig.gca(projection='3d')
ax1.set_title(' Matyas function')
ax1.plot_surface(x, y, z,cmap=cm.nipy_spectral)
plt.show()
```
