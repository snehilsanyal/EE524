```python
##1(a)
import numpy as np
from random import uniform as uni
import matplotlib.pyplot as plt
x = np.linspace(-10,10,100)
# the probabilty function for a normal distribution N(0,1)
pdf2 = (1 / np.sqrt(2 * np.pi * 1)) * (np.exp(-0.5 * ((x - 0) / 1)**2))
plt.plot(x,pdf2,color = 'orange')
plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('Normal distribution mean= 0 and sigma =1')
plt.show()
```

```python
#the probabilty function for a normal distribution N(0,1)
pdf2 = (1 / np.sqrt(2 * np.pi * 4)) * (np.exp(-0.5 * ((x - 3) / 2)**2))
plt.plot(x,pdf2,color = 'orange')
plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('Normal distribution mean= 3 and sigma =2')
plt.show()
```

The differnce between the plots is that the plot has its mean shifted from 0 and has more varying data from the mean.
The magnitude is higher in case of the first plot is because the density of the samples is more at the mean as compared to the second plot.

```python
##1(b)
from mpl_toolkits import mplot3d
x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)
#introducing the necessary vectrs i.e mean and deviation
sigma = np.array([(1,.6),(.6,2)])
#determinant of the sigma vector
d = np.linalg.det(sigma)
#mean vector
u = np.array([(0),(0)])
#transpose(mean)
u = u.reshape(2,1)
#inverse(sigma)
sigmainv = np.linalg.inv(sigma)
#initialisng pdf array
pdf = [[0 for i in range(100)] for j in range(100)]
#nested for loop for obtainin the required pdf array
for i in range(len(x1)):
    for j in range(len(x2)):
        #storin x in t vector
        t = np.array([(x1[i]),(x2[j])])
        #transpose(t)
        t = t.reshape(2,1)
        #necessary vectors to be used in pdf formula
        a = np.transpose(np.subtract(t, u))
        b = np.subtract(t, u)
        c = np.matmul(a,sigmainv)
        #using the pdf definition for multvariate
        pdf1 = (.5 * (np.pi) / d) * np.exp(-.5 * np.matmul(c,b))
        #storing the pdf value in pdf
        pdf[i][j] = float(pdf1)
#creating meshgrid
x1,x2 = np.meshgrid(x1,x2)
pdf = np.array(pdf)
ax = plt.axes(projection = '3d')
ax.plot_surface(x1,x2,pdf)
plt.title('Multivariate Normal distribution')
plt.show()
```
