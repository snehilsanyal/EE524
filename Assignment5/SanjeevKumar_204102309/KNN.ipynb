```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from statistics import mode
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)
M = data.shape[0]
NormalizedData = pd.DataFrame([])
a,b = data.shape
for i in range(b-1):
    NormalizedData[i] = data[i]/(max(data[i])-min(data[i]))
print('NormalizedData\n',NormalizedData.head())
```

```python
# Eucledian distance Function
Origdata = data #savng original data in other variable
data = data.drop([4], axis=1)# dropping label column


def Edist(arr1,arr2):
    d = 0
    for i in range(len(arr1)):
        d = d + (arr1[i]-arr2[i])**2

    return np.sqrt(d)
```

```python
# Define Nearest neighour function that returns the least distances
def knn(data,test,k):

    d = []
    knns = []
    for i in range(M):
        x = Edist(data.iloc[i],test)
        d.append(x)
    while len(knns)<k:
        t = min(d)

        knns.append(t)
        d.remove(t)
    return knns
```

```python
# Classification Solutions returning the most frequent classes of the k nearest points
def prediction(data,test,k):

    d = []
    li = []
    nearest=[]
    for i in range(M):
        x = Edist(data.iloc[i], test)
        d.append(x)
    for i in range(k):
        t = min(d)
        g = d.index(t)
        li.append(g+i)
        d.remove(t)
    for l in range(len(li)):
        nearest.append(Origdata.iloc[li[l],4])
    return mode(nearest)
```

```python
# Accuracy test

classes = list(Origdata[4])
predicted =[]
def Accuracy(k):
    count = 0
    for i in range(M):
        if classes[i] == prediction(data,data.iloc[i],k):
            count = count+1
    print('Accuracy when k is:',k)
    return count/M
```
