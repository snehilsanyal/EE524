```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('File.csv')
df.head()
```

```python
#plotting the all points in data set scatter form before applying K-mean..


plt.scatter(df.feature1,df['feature2'])
plt.xlabel('feature1')
plt.ylabel('feature2')
```

```python
#Applying K-mean cluster with K=2

km = KMeans(n_clusters=2)
y_predicted = km.fit_predict(df[['feature1','feature2']])
df['cluster'] = y_predicted
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
plt.scatter(df1.feature1,df1['feature2'],color='green')
plt.scatter(df2.feature1,df2['feature2'],color='red')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('feature1')
plt.ylabel('feature2')
plt.legend()
```
