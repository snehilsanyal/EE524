# Assignment 7 : Support Vector Machine Using SciKit Learn

```python
import pandas as pnd
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
def iris_binarylabels(irislabels):
    
    x=[]
    for i in range(len(irislabels)):
        if irislabels[i]=='Iris-setosa':
            x.append(1)
        elif irislabels[i]=='Iris-versicolor':
            x.append(2)
        else:
            x.append(3)
    return x
```

Q1: Load the dataset

```python
rawirisdata=pnd.read_csv(r"C:\Users\keyen\Downloads\iris.data", sep=',', names=["SepLen","SepWid","PetLen","PetWid","Species"])
irisdata = rawirisdata.drop('Species', axis=1)
irislabels = rawirisdata['Species']
```

 Q2: Split the dataset

```python
#Split the dataset for training
traindata, testdata, trainlabel, testlabel = train_test_split(irisdata, irislabels, test_size = 0.25)
```

 Q3 : Train the dataset

```python
svclassifier = SVC(kernel='linear');
svclassifier.fit(traindata, trainlabel)
```

Q4 : Test the datset

```python
prediction = svclassifier.predict(testdata)
```

Q5: Permormance of model

```python
confusionmatrix=confusion_matrix(testlabel,prediction)
fig=plt.figure()
ax=plt.gca()
sn.heatmap(confusionmatrix, annot=True,cmap='Blues',yticklabels=['Iris-setosa','Iris-versicolor','Iris-virginica']
           ,xticklabels=['Iris-setosa','Iris-versicolor','Iris-virginica'])
plt.xlabel("Predicted Class",labelpad=20)
plt.ylabel("Actual Class",labelpad=20)
plt.title('Confusion Matrix')
plt.yticks(rotation='horizontal')
plt.show()
```

 Q6 : Accuracy of the classification

```python
print(classification_report(testlabel,prediction))
```

```python
binarylabel=iris_binarylabels(prediction)

fig3=plt.figure()
ax3=plt.gca()
testdata.plot.scatter(x='PetLen',y='PetWid',c=binarylabel,colormap='viridis',ax=ax3,edgecolor='k',style='o',s=40,legend=True)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Plot with Predicted Labels')
plt.show()
```

 Q7: Classification with 80% training data and 20% testing data

```python
traindata, testdata, trainlabel, testlabel = train_test_split(irisdata, irislabels, test_size = 0.20)
svclassifier = SVC(kernel='linear')
svclassifier.fit(traindata, trainlabel)
prediction = svclassifier.predict(testdata)
```

```python
confusionmatrix=confusion_matrix(testlabel,prediction)
fig=plt.figure()
ax=plt.gca()
sn.heatmap(confusionmatrix, annot=True,cmap='BuGn',yticklabels=['Iris-setosa','Iris-versicolor','Iris-virginica']
           ,xticklabels=['Iris-setosa','Iris-versicolor','Iris-virginica'])
plt.xlabel("Predicted Class",labelpad=20)
plt.ylabel("Actual Class",labelpad=20)
plt.title('Confusion Matrix')
plt.yticks(rotation='horizontal')
plt.show()
```
